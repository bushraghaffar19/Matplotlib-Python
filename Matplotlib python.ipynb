# Matplotlib - Data Visualization

```python
import matplotlib.pyplot as plt
x=[12,25,56,62]
y=[1.2,1.5,1.6,1.8]
plt.plot(x,y,color = 'red')
plt.show()
```

```python
#Represent data in form of bar graph
import matplotlib.pyplot as plt
x=[12,25,56,62]
y=[1.2,1.5,1.6,1.8]
plt.bar(x,y)
plt.show()
```

```python
#Represent data in form of scattered Graph
import matplotlib.pyplot as plt
x=[12,25,56,62]
y=[1.2,1.5,1.6,1.8]
plt.scatter(x,y)
plt.show()
```

```python
#Represent data in form of pie Graph
import matplotlib.pyplot as plt
x=[12,25,56,62]
y=[1.2,1.5,1.6,1.8]
plt.pie(x,y)
plt.show()
```

```python
#Represent data in form of Histogram
import matplotlib.pyplot as plt
x=[1,1.2,2,3.2,3.6,3.65,3.4,3.9,3.88,4,5.2,5.4,6]
plt.hist(x,bins=6)
plt.show()
```

```python
# Import numpy as np
import numpy as np

# Import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

pop=[256,236,245,654,785,985,265,212]

# Store pop as a numpy array: np_pop
np_pop=np.array(pop)

# Double np_pop
np_pop=np_pop*2

# Update: set s argument to np_pop
gdp_cap = [1002,1023,1056,10096,10085,10075,100256,100587]
life_exp = [42,52,45,65,80,75,46,67]

plt.scatter(gdp_cap, life_exp, s = np_pop)

# Previous customizations
plt.xscale('log') 
plt.xlabel('GDP per Capita [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007')

# Display the plot
plt.show()
```

```python
# Import numpy as np
import numpy as np

# Import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

pop=[256,236,245,654,785,985,265,212]

# Store pop as a numpy array: np_pop
np_pop=np.array(pop)

# Double np_pop
np_pop=np_pop*2

# Update: set s argument to np_pop
gdp_cap = [1002,1023,1056,10096,10085,10075,100256,100587]
life_exp = [42,52,45,65,80,75,46,67]

plt.scatter(gdp_cap, life_exp, s = np_pop)

# Previous customizations
plt.xscale('log') 
plt.xlabel('GDP per Capita [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007')
plt.text(10096,65,'Pakistan')
plt.text(1135,45,'India')
plt.grid(True)

# Display the plot
plt.show()
```
